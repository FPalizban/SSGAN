```python
import pandas as pd
import numpy as np
import keras
import os
```

```python
!cp /content/gdrive/MyDrive/Code/SSGAN/preprocessing.py /content/preprocessing.py

import preprocessing
```

```python
root = 'gdrive/My Drive/SSGAN'
```

```python
unlabel_vcf = pd.read_excel(os.path.join(root,'Example_Classification/unlabel_data.xlsx'), sheet_name=0, header=0)
unlabel_seq = open(os.path.join(root,'Example_Classification/unlabel_data_seq.txt') , 'r').readlines()
```

```python
unlabel = preprocessing.initial_process(unlabel_vcf, unlabel_seq)
```

```python
Classifier = keras.models.load_model(os.path.join(root, 'model_weights/classifier.keras'))
```

```python
y_pred = Classifier.predict(unlabel)

y_pred = [1 if num>0.5 else 0 for num in y_pred]
```

```python
Labeled = pd.DataFrame()
Labeled['chr'] = unlabel['Chromosome']
Labeled['Start_Position'] = (unlabel['Start_Position']*120000000)+100000000
Labeled['Start_Position'] = Labeled['Start_Position'].apply(int)
Labeled['Ref'] = unlabel['Ref'].replace([0,1,2,3], ['A','T','C','G'])
Labeled['Alt'] = unlabel['Alt'].replace([0,1,2,3], ['A','T','C','G'])
Labeled['Label'] = y_pred
Labeled['Label'].replace([0,1], ['tissue','CH'], inplace=True)
Labeled[90:100]
Labeled.to_csv(os.path.join(root, 'Example_Classification/predicted_labels.csv'))
```
