```python
import pandas as pd
import numpy as np
import keras
import os
```

```python
root = 'gdrive/My Drive/SSGAN'
```

```python
generator = keras.models.load_model(os.path.join(root, 'model_weights/generator.keras'))
```

```python
X_fake, _ = SSGAN_model.generate_fake_samples(generator, 50, 1000)
```

```python
Column_list = ['Chromosome','Start_Position','Alt','Ref','1 position',
               '-1 position','2 position','-2 position','3 position','-3 position']

X_fake = pd.DataFrame(X_fake, columns=Column_list)
```

```python
X_fake['Chromosome'] = round(X_fake['Chromosome'])

X_fake['Start_Position'] = X_fake['Start_Position']*120000000 + 100000000
X_fake['Start_Position'] = X_fake['Start_Position'].apply(int)

for i in Column_list:
  X_fake[i]=round(X_fake[i])

Column_N = ['Alt',	'Ref',	'1 position',	'-1 position',	'2 position',	'-2 position',	'3 position',	'-3 position']
for i in Column_N:
  X_fake[i].replace([0,1,2,3,4], ['A','T','C','G','G'], inplace=True)
```
